```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

# 1. Get Data

```python
solar_data = pd.read_csv('Copy of sonar data.csv',header=None)
solar_data.head()
```

# 2. Preprocess Data

```python
solar_data.shape
```

```python
solar_data.duplicated().sum()
```

```python
solar_data[60].value_counts()
```

```python
solar_data.groupby(60).mean()
```

```python
solar_data.describe()
```

# 3. Train Test Split

```python
X = solar_data.drop(columns=60,axis=1)
y = solar_data[60]
```

```python
print(X.shape)
print(y.shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1,stratify=y, random_state=1)
```

```python
print(X_train.shape,X_test.shape,'\n')
print(y_train.shape,y_test.shape)
```

# 4. Build and Evaluate Model

```python
model = LogisticRegression()
model.fit(X_train,y_train)
```

```python
training_prediction = model.predict(X_train)
print(accuracy_score(training_prediction,y_train))
```

```python
test_prediction = model.predict(X_test)
print(accuracy_score(test_prediction,y_test))
```
